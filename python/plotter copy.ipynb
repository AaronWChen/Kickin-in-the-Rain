## Import necessary libraries

```python
import pandas as pd
import sqlite3
import numpy as np
from WeatherGetter import WeatherGetter as wg
import matplotlib.pyplot as plt
%matplotlib inline
```

## Import Combined Data

```python
df = pd.read_csv('../data/2011_summary.csv')
```

```python
f, ax = plt.subplots(1, figsize=(20,10))
bar_width = 1
bar_l = [i for i in range(len(df['Club']))]
tick_pos = [i + (bar_width/15) for i in bar_l]

p1 = ax.bar(bar_l, df['wins'], label='Wins', alpha=0.9, color='#40F647',
        width=bar_width, edgecolor='white')

p2 = ax.bar(bar_l, df['draws'], bottom=df['wins'], label='Draws',
        alpha=0.9, color='#cc00cc',width=bar_width,edgecolor='white')

p3 = ax.bar(bar_l,df['losses'],bottom=df['wins'] + df['draws'], label='Losses',
        alpha=0.9, color='#4166f5', width=bar_width, edgecolor='white')

plt.xticks(tick_pos, df['Club'])
ax.set_ylabel("Match Results", fontsize=24)
ax.set_xlabel("Clubs", fontsize=24)

plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.ylim(0,40)
ax.legend([p1, p2, p3], ['Wins', 'Draws', 'Losses'], fontsize=14)

plt.setp(plt.gca().get_xticklabels(),rotation=45,horizontalalignment='right',fontsize=12)
# plt.savefig('../plots/2011_season_wins_draws_losses_summary.png')
plt.show()
```

```python
f, ax = plt.subplots(1, figsize=(20,10))
bar_width = 1
bar_l = [i for i in range(len(df['Club']))]
tick_pos = [i + (bar_width/15) for i in bar_l]

p1 = ax.bar(bar_l, df['rain_win_percent'], label='Wins', alpha=0.9, color='#4166f5',
        width=bar_width, edgecolor='white')

plt.xticks(tick_pos, df['Club'])
ax.set_ylabel("Winning Percentage in Rain", fontsize=24)
ax.set_xlabel("Clubs", fontsize=24)

plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.ylim(0,105)

plt.setp(plt.gca().get_xticklabels(),rotation=45,horizontalalignment='right', fontsize=12)
# plt.savefig('../plots/2011_season_rain_win_percentage_summary.png')
plt.show()
```
